```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
```

```python
ftData = pd.read_csv("FullTimeEmployeeData.csv")
ftData.head()
```

```python
protocol = 'postgresql'
username = 'postgres'
password = 'postG'
host = 'localhost'
port = 5432
database_name = 'DataScienceSalaries'
rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string)
```

```python
# check for table names
engine.table_names()
```

```python
# load csv
ftData.to_sql(name="salaries", con=engine, if_exists='append', index=False)
```

```python
# check
pd.read_sql_query('select * from salaries', con=engine).head()
```
